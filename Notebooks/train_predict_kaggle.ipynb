```python
import os
import shutil

working_dir = "/kaggle/working"

# Supprimer tous les fichiers et dossiers dans working
for filename in os.listdir(working_dir):
    file_path = os.path.join(working_dir, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)  # supprime les fichiers ou liens symboliques
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)  # supprime les dossiers
    except Exception as e:
        print(f"Erreur lors de la suppression de {file_path}: {e}")

print("Le dossier working est maintenant vide !")
```
